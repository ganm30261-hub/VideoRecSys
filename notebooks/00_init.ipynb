# 🚀 VideoRecSys — 标准初始化 Cell
**每次打开 Colab 第一件事：运行下面的初始化 Cell**

完成以下4步：
```
① 挂载 Google Drive（数据永久存储，不怕 Session 重置）
② 从 GitHub 拉取最新代码
③ 创建目录结构
④ 验证环境 + 检查 Step 1 数据是否存在
```

````python
# ============================================================
# 🚀 VideoRecSys 标准初始化 Cell
# 复制到每个 Notebook 的最顶部
# 每次打开 Colab 只运行这一个 Cell 即可
# ============================================================
import os
from google.colab import drive, userdata

# ─── ① 挂载 Google Drive ───
print('① 挂载 Google Drive...')
drive.mount('/content/drive')
print('   ✅ Google Drive 挂载完成')

# ─── ② 从 GitHub 拉取最新代码 ───
print('\n② 同步 GitHub 代码...')
TOKEN = userdata.get('GITHUB_TOKEN')
REPO  = 'ganm30261-hub/VideoRecSys'
LOCAL = '/content/VideoRecSys'

if not os.path.exists(f'{LOCAL}/.git'):
    os.system(f'rm -rf {LOCAL}')
    os.system(f'git clone https://{TOKEN}@github.com/{REPO}.git {LOCAL} -q')
    print('   ✅ 仓库克隆完成')
else:
    os.chdir(LOCAL)
    os.system(f'git remote set-url origin https://{TOKEN}@github.com/{REPO}.git')
    os.system('git pull origin main -q')
    print('   ✅ 代码已更新到最新')

os.chdir(LOCAL)
print('   最近提交: ', end='')
os.system('git log --oneline -1')

# ─── ③ 创建目录结构（指向 Google Drive）───
print('\n③ 创建目录结构...')
BASE_DIR = '/content/drive/MyDrive/VideoRecSys'
PATHS = {
    'raw'          : f'{BASE_DIR}/data/raw',
    'cleaned'      : f'{BASE_DIR}/data/cleaned',
    'features'     : f'{BASE_DIR}/data/features',
    'samples'      : f'{BASE_DIR}/data/samples',
    'feature_store': f'{BASE_DIR}/feature_store',
    'models_recall': f'{BASE_DIR}/models/recall',
    'models_rank'  : f'{BASE_DIR}/models/ranking',
    'logs'         : f'{BASE_DIR}/logs',
    'reports'      : f'{BASE_DIR}/reports',
}
for p in PATHS.values():
    os.makedirs(p, exist_ok=True)
print('   ✅ 目录结构就绪')
print(f'   数据根目录: {BASE_DIR}')

# ─── ④ 验证环境 ───
print('\n④ 验证环境...')
import torch, pandas as pd, numpy as np

gpu = torch.cuda.get_device_name(0) if torch.cuda.is_available() else '无 GPU（建议开启 T4）'
print(f'   GPU:     {gpu}')
print(f'   PyTorch: {torch.__version__}')
print(f'   Pandas:  {pd.__version__}')

# 检查 Step 1 数据
step1_files = [
    (f"{PATHS['samples']}/train.parquet",           'train.parquet'),
    (f"{PATHS['samples']}/test.parquet",            'test.parquet'),
    (f"{PATHS['features']}/user_features.parquet", 'user_features.parquet'),
    (f"{PATHS['features']}/item_features.parquet", 'item_features.parquet'),
    (f"{PATHS['features']}/user_sequences.parquet",'user_sequences.parquet'),
]
print('\n   Step 1 数据检查:')
all_ready = True
for fp, name in step1_files:
    if os.path.exists(fp):
        mb = os.path.getsize(fp) / 1024**2
        print(f'   ✅ {name:35s} {mb:.1f} MB')
    else:
        print(f'   ❌ {name:35s} 不存在 → 需要先跑 Step 1')
        all_ready = False

print('\n' + '='*50)
if all_ready:
    print('  🎉 环境就绪！直接运行当前 Notebook')
else:
    print('  ⚠️  请先运行 step1_data_pipeline.ipynb')
print('='*50)
````
